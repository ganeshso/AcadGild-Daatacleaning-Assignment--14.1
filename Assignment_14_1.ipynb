```python
import pandas as pd
import sqlite3 as sql
from sqlalchemy import create_engine
```

## 1. Create an sqlalchemy engine using a sample from the data set

```python
engine = create_engine('sqlite:///sqladb', echo=True)
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',header=None)
df.columns =['age','workclass','fnlwgt','education','education-num','marital-status','occupation',
           'relationship','race','sex','capital-gain','capital-loss',
           'hours-per-week','native-country','salary_slab']
df.to_sql('sqladb',engine,if_exists='replace')
df.head()
```

```python
connection = engine.connect()
```

```python
pd.read_sql_query("select * from sqladb limit 20",con=connection)
```

## 5. Write two function queries

```python
def getDistribution(columnName):
    result = pd.read_sql_query("Select "+columnName+", count(*) as Frequency from sqladb group by " + columnName,con=connection)
    return result


getDistribution('age')
getDistribution('sex')
```

```python
def insertNewRecord(dbName, record_data):
    values   = ',?' * (len(record_data)-1)
    query = "insert into " + dbName + " values(?"+ values +")"
    connection.execute(query,list(record_data))
    
insertNewRecord('sqladb',('99999','99','Self-emp','26716','Bachelors','25','Married-civ-spouse','Exec-managerial','Husband','White','Male','0','0','20','USA_22','>50K'))


                
```
